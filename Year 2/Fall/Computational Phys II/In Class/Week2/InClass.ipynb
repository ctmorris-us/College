```python
import math
import numpy as np
import matplotlib.pyplot as plt

delta = 0.0
alpha = -2.0
beta = 1.0

def deriv2(x, y):
    return -delta*y[1] - alpha*y[0] - beta*y[0]**3

def phi(x, y):
    return 0.5*alpha*y[0]**2 + 0.25*beta*y[0]**4

def energy(x, y):
    return 0.5*y[1]**2 + phi(x, y)

def f(x, y):
    return np.array([y[1], deriv2(x, y)])

def euler_step(func, x, y, dx):
    y += dx*func(x, y)
    x += dx
    return x, y

def midpoint_step(func, x, y, dx):
    dy = dx*func(x, y)
    y += dx*func(x+0.5*dx, y+0.5*dy)
    x += dx
    return x, y

def get_y_vy(x0, y0, vy0, dx, xmax):

    xplots = []
    yplots = []
    vplots = []
    eplots = []

    x = x0
    y0 = y0
    v0 = vy0

    y = np.array([y0, v0])
    xplot = [x]
    yplot = [y[0]]
    vplot = [y[1]]
    eplot = [energy(x, y)]

    while x < xmax:
        yp = y[0]
        vp = y[1]
        if x == 0: yp = y0 + 1
        x, y = midpoint_step(f, x, y, dx)
        xplot.append(x)
        yplot.append(y[0])
        vplot.append(y[1])
        eplot.append(energy(x, y))
        if yp <= y0 and y[0] > y0:
            break

    vint = vp + (y[1]-vp)*(y[0]-y0)/(y[0]-yp)
    error = abs((vint-v0)/v0)
    print('vint =', vint, 'err =', error)
    
    return error, yplot, vplot

x = 0
y0 = 1.3
vy0 = 1.5
xmax = 10.*math.pi
dx = 0.00009


v_possible_initials = [.25, .5, 1, 1.25, 1.5, 2]

fig, ax = plt.subplots(2, 1, figsize = (7, 7))

for v_initial in v_possible_initials:
    error, yplot, vplot = get_y_vy(x, y0, v_initial, dx, xmax)

    ax[0].plot(yplot, vplot)
    ax[0].set(xlabel = 'y', ylabel = 'vy', xlim = (-2, 2))
    
    ax[1].plot(v_initial, error, 'ko', markersize = 5)
    
```

```python
import math
import numpy as np
import matplotlib.pyplot as plt

delta = 0.0
alpha = -2.0
beta = 1.0

def deriv2(x, y):
    return -delta*y[1] - alpha*y[0] - beta*y[0]**3

def phi(x, y):
    return 0.5*alpha*y[0]**2 + 0.25*beta*y[0]**4

def energy(x, y):
    return 0.5*y[1]**2 + phi(x, y)

def f(x, y):
    return np.array([y[1], deriv2(x, y)])

def euler_step(func, x, y, dx):
    y += dx*func(x, y)
    x += dx
    return x, y


def rk4_step(func, x, y, dx):
    dy1 = dx*func(x, y)
    dy2 = dx*func(x+0.5*dx, y+0.5*dy1)
    dy3 = dx*func(x+0.5*dx, y+0.5*dy2)
    dy4 = dx*func(x+dx, y+dy3)
    y += (dy1 + 2*dy2 + 2*dy3 + dy4)/6.
    x += dx
    return x, y

xmax = 10.*math.pi
dx = 0.01
period_plot = []
vy0_plot = []

fig, ax = plt.subplots(2, 1, figsize = (7, 7))

for v0 in np.arange(.1, 10.01, .01):
    
    xplots = []
    yplots = []
    vplots = []
    eplots = []

    x = 0.0
    y0 = 1.3

    y = np.array([y0, v0])
    xplot = [x]
    yplot = [y[0]]
    vplot = [y[1]]
    eplot = [energy(x, y)]
    
    while x < xmax:
        yp = y[0]
        vp = y[1]
        if x == 0: yp = y0 + 1
        x, y = rk4_step(f, x, y, dx)
        xplot.append(x)
        yplot.append(y[0])
        vplot.append(y[1])
        eplot.append(energy(x, y))
        if yp <= y0 and y[0] > y0:
            period_plot.append(x)
            vy0_plot.append(v0)
            break

    vint = vp + (y[1]-vp)*(y[0]-y0)/(y[0]-yp)
    print('v0 =', v0, 'vint =', vint,
          'err =', abs((vint-v0)/v0))

    ax[0].plot(yplot, vplot)
    
ax[0].set(xlabel = 'y', ylabel = 'vy')
    
ax[1].plot(vy0_plot, period_plot)
ax[1].set(xlabel = 'vy0', ylabel = 'period', xlim = (0, 10.01))


```
