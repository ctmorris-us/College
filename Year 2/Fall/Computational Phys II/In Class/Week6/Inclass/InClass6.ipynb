```python
import math
import numpy as np
from scipy.linalg import solve, det
import matplotlib.pyplot as plt

def g(x):
    return 1

a = 0
b = 1.0
ya = 1
yb = 2.0
N = 1000

delta = (b-a)/N
x = np.linspace(a, b, N+1)

# Simple difference matrix A (dimension N-1).

xx = x[1:-1]		# N-1 interior points only: indices n = 0 to N-2

A = np.zeros((N-1,N-1))
for n in range(N-1):
    A[n,n] = -2.0 + delta**2*g(xx[n])
    if n > 0:
        A[n,n-1] = 1.0
    if n < N-2:
        A[n,n+1] = 1.0

# Right-hand side r.
        
r = np.zeros(N-1)
r[0] = -ya
r[-1] = -yb

print(A)
print(r)

y0 = solve(A, r)
print (y0)
```

```python
plt.plot(xx, y0, 'k-')

def sol(x):
    B = (2 - np.cos(1))/np.sin(1)
    return np.cos(x) + B*np.sin(x)

plt.plot(xx, sol(xx),'b-.')
```
