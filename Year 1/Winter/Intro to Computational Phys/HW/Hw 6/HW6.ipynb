Program that is made to plot the trajetory of a particle under the influece of air resistnace across a non-curvature plane. The program finds the projectiles time of flight, range of flight, and angle of collision. The program also finds the initial velocity that is required for the particle to match the range of flight had it not been slowed by air resistance. The program also varies the angle of launch to find the angle that maximizes the range of flight for the particle under air resistance.

#### !!! The angle of collision is measured from theta = 0 counterclockwise !!!

```python
import math
import matplotlib.pyplot as plt

beta_ls = [.001]
range_of_flight_ls = []
time_of_flight_ls = []
for beta_int in beta_ls:
    
# acceleration due to gravity
    g = 9.8
    alpha = 0
    beta = beta_int

    height = []
    pos_height = []


    def interp(x0, y0, x1, y1, x=None, y=None):
        if y == None:
            return y0 + (y1 - y0) * (x - x0) / (x1 - x0)
        elif x == None:
            return (y-y0) * (x1 - x0) / (y1 - y0) + x0

    def acc(x, y, vx, vy, t):
        global alpha, beta
        
        v_abs = math.sqrt(vx**2 + vy**2)
        return - beta*v_abs*vx, -g - beta*v_abs*vy
    

    def take_a_step(x, y, vx, vy, t, dt):

        ax,ay = acc(x, y, vx, vy, t)

        # Predict: 
        x  += vx*dt + 0.5*ax*dt*dt
        y  += vy*dt + 0.5*ay*dt*dt
        vx += ax*dt
        vy += ay*dt

        # Correct: 
        ax1,ay1 = acc(x, y, vx, vy, t)

        vx += 0.5*(ax1-ax)*dt
        vy += 0.5*(ay1-ay)*dt

        t += dt

        return x,y,vx,vy,t


    def check_amplitude(yp, y, yf, x):
            if (yp < y) and (yf <= y):
                height.append(y)
                pos_height.append(x)


    # Set initial position and velocity.

    x0 = 0.0
    y0 = 0.0

    v0    = 100				# unit: m/s 
    theta = 60.0				# (degrees) , angle with horizontal 

    # Determine components of the initial velocity vector.
    vx0 = v0 * math.cos( math.radians(theta) )
    vy0 = v0 * math.sin( math.radians(theta) )

    # Set parameters governing the numerical details.
    dt    = 0.01
    t_max = 15.0

    # Initialize the trajectory.
    t = 0.0

    x = x0
    xp = x
    xpp = xp

    y = y0
    yp = y
    ypp = yp

    vx = vx0
    vy = vy0
    
    x_pos = [x0]
    y_pos = [y0]
    x_vel = [vx0]
    y_vel = [vy0]
    
    time = [0]


    # Calculate the trajectory to time t_max, using the 2D
    # predictor-corrector scheme.

    while y>=0:
        xpp, ypp = xp, yp
        xp, yp = x, y
        x,y,vx,vy,t = take_a_step(x, y, vx, vy, t, dt)
        
#         xf, yf = take_a_step(x, y, vx, vy, t, dt)[0:2]
        check_amplitude(ypp, yp, y, x)

        x_pos.append(x)
        y_pos.append(y)
        x_vel.append(vx)
        y_vel.append(vy)
        time.append(t)

    
    #plt.plot(pos_height, height, 'o')

    x_of_flight  = interp(x_pos[-2], y_pos[-2], x_pos[-1], y_pos[-1], y = 0)   
    vx_of_flight = interp(x_vel[-2], y_pos[-2], x_vel[-1], y_pos[-1], y = 0)   
    vy_of_flight = interp(y_vel[-2], y_pos[-2], y_vel[-1], y_pos[-1], y = 0)
    
    v_of_flight = math.sqrt(vx_of_flight**2 + vy_of_flight**2)
    angle_of_collision = math.atan( (-vy_of_flight/v_of_flight) / (vx_of_flight/v_of_flight) )
    angle_of_collision = 180 - math.degrees(angle_of_collision)
    
    time_of_flight  = interp(time[-2], y_pos[-2], time[-1], y_pos[-1], y = 0)
    range_of_flight_ls.append(x_of_flight)
    time_of_flight_ls.append(time_of_flight)
    
    plt.plot(x_pos, y_pos, label = 'Trajectory')
    plt.plot((x_of_flight, vx_of_flight*math.cos(math.radians(angle_of_collision)) + x_of_flight), 
             (0, -vy_of_flight*math.sin(math.radians(angle_of_collision))), label = 'Tangent Vector')
   
    plt.title('a) x vs y')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    
    #X_of_flight is the range of flight because it's the distance traveled
    print('a)  ROF = {:.5f} m , TOF = {:.5f} s, beta = {}'.format(x_of_flight, time_of_flight, beta))
    print('b1) With Air Resistance')
    print('b1) Angle with the ground when the object strikes the ground: {} degrees'.format(angle_of_collision))


```
