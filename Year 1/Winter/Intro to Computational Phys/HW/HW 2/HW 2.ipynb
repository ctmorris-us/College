1)

(a) Find numerically all maxima, minima, and zeros (x-values) of the Bessel func- tions J0(x), J1(x), and J2(x) in the range 0 ≤ x ≤ 30. Grid the functions with an interval dx = 0.1.

(b)  Use linear interpolation to refine your estimates of the zeros (you do not need to use quadratic interpolation for the maxima or minima). As discussed in class, the Bessel functions are accessible as jn(0,x), jn(1,x), and jn(2,x).


Comments:
The program runs three times each for the different bessel functions (n=0, 1, and 2). The program makes empty lists for holding the maxs, mins, and zeros. The program makes 3 variables for x and y, a past, current, and future variable. This is done to check the neighborhood around the current x or y and establish if it's a max, min, or zero. The program checks for maxs by checking if the current x is greater than the past x and greater than or equal to the future x. The program checks for mins by checking if the current x is less than the past x and less than or equal to the future x. The program checks for zeros by either checking if the past x is below zero and the future x is above 0 or if the past x is above zero and the future x is below zero. The first graph has multiple overlapping zeros because the time step is so large that the zero checker finds a zero twice once for each timestep.



```python
import matplotlib.pyplot as plt
from scipy.special import jn
import numpy as np

#Self made interpolation function that allows for either interpolated x or y
def interp(x0, y0, x1, y1, x=None, y=None):
    if y == None:
        return y0 + (y1 - y0) * (x - x0) / (x1 - x0)
    elif x == None:
        return (y-y0) * (x1 - x0) / (y1 - y0) + x0


# Done each time for j0, j1, j2    
for i in range(0,3):
    dx = .05
    

    # Initializes x,y and future x,y
    x = 0 
    xf = dx 
    
    y = jn(i, x) 
    yf  = jn(i, xf)
    
    x_list =[]
    y_list =[]
    x_max_list  = []
    y_max_list  = []
    x_min_list  = []
    y_min_list  = []
    x_zero_list = []
    y_zero_list = []
    x_zero_list_interpolated = []
    y_zero_list_interpolated = []
    line_styles = ['-', '--', '-.'] #list to reference when plotting

    while x<=30:
        # Add current values of x and y to lists.
        x_list.append(x)
        y_list.append(y)
        
        # Reestablishes the past x,y; current x,y; and future x,y for determining maxs, mins, and zeros
        xp = x
        x  = xf
        xf += dx
        
        yp = y
        y  = yf
        yf = jn(i, xf)
        
        if (y > yp) and (y >= yf): #Checks for max by checking if current value is greater than the past and future.
            x_max_list.append(x)
            y_max_list.append(y)
        
        if (y < yp) and (y <= yf): #Checks for min by checking if current value is less than the past and future.
            x_min_list.append(x)
            y_min_list.append(y)
        
        if ((yp <= 0) and (yf >= 0)) or ((yp >= 0) and (yf <= 0)): #Checks for zero coming from above or below
            x_zero_list.append(x)
            y_zero_list.append(y)
            x_zero_list_interpolated.append(interp(xp, yp, xf, yf, y=0)) #Calls Interpolation Function
            y_zero_list_interpolated.append(0)
    
    #Printing out the maxs, mins, and zeros.
    print('-' * 25)
    print('The Maximum for Bessel n={}: {}\n'.format(i, y_max_list ))
    print('The Minimum for Bessel n={}: {}\n'.format(i, y_min_list ))
    print('The Non-Interpolated Zeros for Bessel n={}: {}\n'.format(i, y_zero_list ))
    print('The Interpolated Zeros for Bessel n={}: {}\n'.format(i, y_zero_list_interpolated ))
    print('-' * 25 + '\n')
    
    # A) Plot non-interpolated zeros for the bessel functions
    plt.figure(1, figsize = (10,10))
    plt.subplot(2,1,1)
    plt.title('A) Bessel Functions (n = 0, 1, and 2) as Functions of x')
    plt.xlabel('x')
    plt.ylabel('J(x)')
    plt.plot(x_list, y_list, 'k{}'.format(line_styles[i]), label = 'J {}'.format(i))
    plt.plot(x_max_list, y_max_list, 'ro', label = 'Maxmium')
    plt.plot(x_min_list, y_min_list, 'bo', label = 'Minimum')
    plt.plot(x_zero_list, y_zero_list, 'go', label = 'Non-Interpolated Zeros')
    
    plt.legend()

    # B) Plot interpolated zeros bessel for the functions
    plt.subplot(2,1,2)
    plt.title('B) Bessel Functions (n = 0, 1, and 2) as Functions of x')
    plt.xlabel('x')
    plt.ylabel('J(x)')
    plt.plot(x_list, y_list, 'k{}'.format(line_styles[i]), label = 'J {}'.format(i))
    plt.plot(x_zero_list_interpolated, y_zero_list_interpolated, 'ro', label = 'Interpolated Zeros')
    
    plt.legend()

```

(a) Use the trapezoid rule to compute and plot the position of the particle

(b) Use numerical differentiation to compute and plot the particle’s acceleration

Comments: The program is relatively small. Plotting j0 is relatively straightforward and is done by calculating j0 for each time step. Calculating the area is straightfoward as well because it's based on a constantly increasing upperbound. The area is done using the trapezoidal method. A global area variable is added onto by one new trapezoid calculated for t to t+dt for each time step. The area (x(t)) is initialized as negative because the final timestep in the while loop adds on an extra trapezoid from the 30 to 30 + dt. The derivative (a(t)) is calculated symmetrically using t+dt and t-dt. Finally, v(t), x(t) are plotted on the first graph and v(t), x(t), and a(t) are plotted on the second graph.


```python
# Establishing dt and t
dt = .1
t = 0
N = 30/dt + 1

t_list =[]
v_t = []
a_t = []
x_t = []

# area is initially negative to compensate for the extra area found from the upperbound to the upperbound + dt.
area = -.5 * (jn(0, 30) + jn(0, 30 + dt)) * dt

while t<=30:
    t_list.append(t) #t value 
    v_t.append(jn(0, t)) #bessel function n=0 for t value.
    
    area += .5 * (jn(0, t) + jn(0, t+dt)) * dt #updates area by adding on the next trapezoid.
    x_t.append(area) 
    a_t.append((jn(0, t+dt) - jn(0, t-dt)) / (2*dt)) #calculates the symmetric derivative of v(t) at a point.
    
    t += dt
    
print('The final v(t) from 0 to 30 is', area)

# A) The first graph is v(t) and x(t) only. 
plt.figure(1, figsize = (10,10))

plt.subplot(2,1,1)
plt.plot(t_list, v_t, 'k-', label = 'v(t)')
plt.plot(t_list, x_t, 'g-.', label = 'x(t)')
plt.xlabel('Time')
plt.ylabel('Bessel function n=0')
plt.title('A) Bessesl Function n=0 and The running Integral as Functions of Time')
plt.legend()

# B) The second graph is v(t), x(t), and a(t).
plt.subplot(2,1,2)
plt.plot(t_list, v_t, 'k-', label = 'v(t)')
plt.plot(t_list, a_t, 'b--', label = 'a(t)')
plt.plot(t_list, x_t, 'r-.', label = 'x(t)')
plt.xlabel('Time')
plt.ylabel('Bessel function n=0')
plt.title('B) The Above Graph plus the Derivative of v(t) as Functions of Time')
plt.legend()
    
            
```

a) Modify the trapezoid integration scheme discussed in class to compute the
work done by the force F on the particle as it moves.

(b) Plot W and x(t) as a function of t for 0≤t≤10.

Comments:
The program has three functions. x_t(t) calculates x(t), F(x) calculates F(x), and W(t) calculates the trapezoid from x(t) to x(t+dt). The program initializes an initial area variable that is negative to compensate for the extra area calculated from tmax to tmax + dt. x(t) and W(t) are plotted from t = 0 to t = 10.

```python
def x_t(t): # Function for calculating x(t)
    return np.exp(-t) * np.sin(2*t)

def F(x): # Function for calculating F(X)
    return -5 * x

# A)
def W(t): # Function for calcuting a trapezoid from x(t) to x(t+dt)
    i1 = x_t(t) #lower bound
    i2 = x_t(t+dt) #upper bound
    return .5*(F(i1) + F(i2))*(i2 - i1)

t_list = []
x_list = []
F_list = []
W_list = []

tmax = 10
dt = .1
t = 0
area = 0 # Done to compensate for the extra trapezoid calculated from tmax to tmax + dt
while t <= (tmax):
    # Finds x(t), F(t), and W(t) for t and adds the trapezoid calculated with W(t) to the total area.
    t_list.append(t)
    x_list.append(x_t(t))
    F_list.append(F(x_t(t)))
    W_list.append(W(t) + W_list[-1])
    area += W(t)
    t += dt
    
print('The Total Work is:',area)

#Plotting x(t) and W(t)
plt.figure(1, figsize = (7,7))
plt.plot(t_list, x_list, 'b-', label = 'position')
#plt.plot(t_list, F_list, 'k--')
plt.plot(t_list, W_list, 'r-.', label = 'work')
plt.xlabel('Time')
plt.ylabel('Position')
plt.title('B) Position and Work as Function of Time')
plt.legend()
```
