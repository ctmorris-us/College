**Instructions**:

  * Using $\tt dt = 0.05$ and for $t>0$, **find** the first maximum of the function $x(t) = J_0(t^2)$


```python
import matplotlib.pyplot as plt
from scipy.special import jn

def x(t):
    return jn(0, t**2)

def interp(x0, y0, x1, y1, x=None, y=None):
    if y == None:
        return y0 + (y1 - y0) * (x - x0) / (x1 - x0)
    elif x == None:
        return (y-y0) * (x1 - x0) / (y1 - y0) + x0

dt = .05
t = 0
t_list =[]
x_list =[]
count = 0
while t<=10:
    t_list.append(t)
    x_list.append(x(t))
    t += dt
    
    if count != 1:
        if (x(t) > x(t-dt)) and (x(t) > x(t+dt)):
            t_max = t
            count += 1

print('The first non-interpolated maximum is at t = {}'. format(t_max))
        
plt.plot(t_list, x_list, 'k-')
plt.plot(t_max, x(t_max), 'ro', label = 'First Max')
plt.xlabel('t')
plt.ylabel('j(t^2)')
plt.title('t vs j(t^2)')
plt.legend()
```
