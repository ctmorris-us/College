```python
import matplotlib.pyplot as plt
from scipy.special import jn
import numpy as np

def interp(x0, y0, x1, y1, x=None, y=None):
    if y == None:
        return y0 + (y1 - y0) * (x - x0) / (x1 - x0)
    elif x == None:
        return (y-y0) * (x1 - x0) / (y1 - y0) + x0

for i in range(0,3):
    dx = .05
    
    x = 0
    xf = dx
    
    y = jn(i, x)
    yf  = jn(i, xf)
    
    x_list =[]
    y_list =[]
    x_max_list  = []
    y_max_list  = []
    x_min_list  = []
    y_min_list  = []
    x_zero_list = []
    y_zero_list = []
    x_zero_list_interpolated = []
    y_zero_list_interpolated = []
    line_styles = ['-', '--', '-.']

    while x<=30:
        
        x_list.append(x)
        y_list.append(y)
        
        xp = x
        x  = xf
        xf += dx
        
        yp = y
        y  = yf
        yf = jn(i, xf)
        
        if (y > yp) and (y >= yf):
            x_max_list.append(x)
            y_max_list.append(y)
        
        if (y < yp) and (y <= yf):
            x_min_list.append(x)
            y_min_list.append(y)
        
        if ((yp <= 0) and (yf >= 0)) or ((yp >= 0) and (yf <= 0)):
            x_zero_list.append(x)
            y_zero_list.append(y)
            x_zero_list_interpolated.append(interp(xp, yp, xf, yf, y=0))
            y_zero_list_interpolated.append(0)
            
    plt.figure(1, figsize = (10,10))
    plt.subplot(2,1,1)
    plt.title('A)')
    plt.xlabel('x')
    plt.ylabel('J(x)')
    plt.plot(x_list, y_list, 'k{}'.format(line_styles[i]), label = 'J {}'.format(i))
    plt.plot(x_max_list, y_max_list, 'ro')
    plt.plot(x_min_list, y_min_list, 'bo')
    plt.plot(x_zero_list, y_zero_list, 'go')
    
    plt.legend()

    plt.subplot(2,1,2)
    plt.title('B)')
    plt.xlabel('x')
    plt.ylabel('J(x)')
    plt.plot(x_list, y_list, 'k{}'.format(line_styles[i]), label = 'J {}'.format(i))
    plt.plot(x_zero_list_interpolated, y_zero_list_interpolated, 'ro')
    
    plt.legend()

```

```python

dt = .1
t = 0
N = 30/dt + 1

t_list =[]
x_list =[]
sym_div = []

area = -.5 * (jn(0, 30) + jn(0, 30 + dt)) * dt

while t<=30:

    t_list.append(t)
    x_list.append(jn(0, t))
    
    area += .5 * (jn(0, t) + jn(0, t+dt)) * dt
    
    sym_div.append((jn(0, t-1) - jn(0, t+1)) / (2*dt)) 
    
    t += dt
    
    
plt.plot(t_list, x_list, 'k-')
plt.plot(t_list, sym_div, 'k--')
print(area)
    
    
            
```

```python
def x_t(t):
    return np.exp(-t) * np.sin(2*t)

def F(x):
    return -5 * x

def W(t):
    upper = x_t(t)
    dx = abs(upper/.1)
    x = 0
    area = 0
    if upper > 0:
        while x < upper:
            area += .5 * (F(x) + F(x+dx)) * dx
            x += dx
        return area
    elif upper < 0:
        while upper < 0:
            area += .5 * (F(upper) + F(upper+dx)) * dx
            upper += dx
        return -1 * area
    else:
        return 0
    
t_list = []
x_list = []
F_list = []
W_list = []

# W = W()
tmax = 10
dt = .1
t = 0

while t <= (tmax):
    t_list.append(t)
    x_list.append(x_t(t))
    F_list.append(F(x_t(t)))
    W_list.append(W(t))
    t += dt
    
plt.plot(t_list, x_list, 'k-')
#plt.plot(t_list, F_list, 'k--')
plt.plot(t_list, W_list, 'k-.')
```
