Only use the $\tt math$ package (not $\tt numpy$).

* Print, in three columns, $x$, $\sin x$, and
  $\cos x$, for $x = 0,\ldots,2\pi$ in steps of $\pi/500$. 
  
  
* Plot these results: $\sin x$ and $\cos x$ versus $x$, on the same
  graph, using $\tt matplotlib$.

Now using $\tt numpy$,

* Create and plot the same data as above using vectorization. Your
  code for this part should be much shorter than the one above.
  
For both plots, make sure that you put legends to identify each curve and also label the $x$ and $y$ axes.

```python
import math as math
import matplotlib.pyplot as plt 
import numpy as np 

step = math.pi/500
i = 0 
x = []
while i <= 2*math.pi:
    x.append(i)
    i += step

for xval in x:
    print('{:<20} \t\t\t {:<20} \t\t\t {:<20}'.format(xval, math.sin(xval), math.cos(xval)))
    
```

```python

# fig = single_frame(xlabel, ylabel, xsize = 14, ysize = 14)
plt.plot(x, [math.sin(xval) for xval in x], 'g', label = 'sin(x)')
plt.plot(x, [math.cos(xval) for xval in x], 'b', label = 'cos(x)')
plt.legend()
plt.show()
```

Now with Numpy 


```python
x_np = np.arange(0, 2*np.pi + np.pi/500, np.pi/500)
plt.plot(x_np, np.sin(x_np), 'g', label = 'sin(x)')
plt.plot(x_np, np.cos(x_np), 'b', label = 'cos(x)')
plt.legend()


```
